```python
def print_two(a, b):
    print("First:", a)
    print("Second:", b)

print_two(10, 20)
```

```python
def print_args(*args):
    for item in args:
        print(item)

print_args(1, 2, 3, 4, "Hello")
```

```python
def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

print_kwargs(name="Manas", age=21, course="Python")
```

```python
def print_list(lst):
    for item in lst:
        print(item)

print_list([10, 20, 30, 40])
```

```python
def max_of_four(a, b, c, d):
    return max(a, b, c, d)

print(max_of_four(10, 25, 7, 18))
```

```python
def sum_list(lst):
    total = 0
    for num in lst:
        total += num
    return total

print(sum_list([1, 2, 3, 4, 5]))
```

```python
def multiply_list(lst):
    result = 1
    for num in lst:
        result *= num
    return result

print(multiply_list([1, 2, 3, 4]))
```

```python
def check_range(num, start, end):
    if start <= num <= end:
        return True
    else:
        return False

print(check_range(5, 1, 10))
```

```python
def even_odd(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")

even_odd(7)
```
