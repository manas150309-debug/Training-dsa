Write a python program to create and print a dictionary which stores your information. 
(name, age, gender .....)


```python

my_info = {
    "Name": "Manas Gupta",
    "Age": 21,
    "Gender": "male",
    "Course": "Engineering",
    "College": "IIT Madras"
}
print("Personal Information:")
for key, value in my_info.items():
    print(f"{key}: {value}")
```

7. Write a python program to create three dictionaries, then create one dictionary that 
will contain the other three dictionaries. 

```python
dict1 = {
    "Name": "Manas",
    "Age": 21
}

dict2 = {
    "Course": "AI & ML",
    "College": "IIT Madras"
}

dict3 = {
    "City": "Chandrapur",
    "Country": "India"
}

# Create one dictionary containing the above three dictionaries
main_dict = {
    "Personal_Info": dict1,
    "Education_Info": dict2,
    "Location_Info": dict3
}

# Print the final dictionary
print("Combined Dictionary:")
print(main_dict)
```

8. Write a python program to convert two lists into a dictionary in a way that item from 
list1 is the key and item from list2 is the value. 

```python
# Create two lists
list1 = ["Name", "Age", "Course"]
list2 = ["Manas", 21, "AI & ML"]

# Convert lists into a dictionary
result_dict = dict(zip(list1, list2))

# Print the dictionary
print("Resulting Dictionary:")
print(result_dict)
```

9. Write a python program to merge two python dictionaries into one dictionary.

```python
dict1 = {"Name": "Manas", "Age": 21}
dict2 = {"Course": "AI & ML", "College": "IIT Madras"}

merged_dict = dict1 | dict2
print(merged_dict)
```

10. Write a python program to get the key of lowest value from the dictionary. 
sample_dict = { 
'C': 92, 
'Java': 66, 
'Python': 85 
}

```python
# Given dictionary
sample_dict = { 
    'C': 92, 
    'Java': 66, 
    'Python': 85 
}

# Get the key with the lowest value
min_key = min(sample_dict, key=sample_dict.get)

# Print the result
print("Key with the lowest value:", min_key)
```
