```python
def unique_list(lst):
    unique = []
    for i in lst:
        if i not in unique:
            unique.append(i)
    return unique

print(unique_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

```python
def print_even(lst):
    for i in lst:
        if i % 2 == 0:
            print(i, end=" ")

print_even([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def is_palindrome(s):
    if s == s[::-1]:
        return True
    else:
        return False

print(is_palindrome("madam"))
```

```python
def minimum(a, b, c):
    return min(a, b, c)

print(minimum(10, 5, 8))
```

```python
def square_list():
    squares = []
    for i in range(1, 31):
        squares.append(i * i)
    return squares

print(square_list())
```

```python
def count_case(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

count_case("Hello World")
```

```python
def is_pangram(s):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    s = s.lower()

    for ch in alphabet:
        if ch not in s:
            return False
    return True

print(is_pangram("The quick brown fox jumps over the lazy dog"))
```

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False

    for ch in s1:
        if s1.count(ch) != s2.count(ch):
            return False
    return True

print(is_anagram("anagram", "nagaram"))
```
